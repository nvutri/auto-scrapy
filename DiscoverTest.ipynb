```python
def find_repeat(root, xpath):
#     print(root, xpath)
    children_elems = list(filter(lambda x: isinstance(x,html.HtmlElement) , root.getchildren()))
    children_elems = sorted(children_elems, key=lambda elem: elem.tag)
#     print(children_elems)
    for first, second in zip(children_elems, children_elems[1:]):
        if first.tag == second.tag and first.classes == second.classes:
            new_xpath = '%s.%s' % (first.tag, first.classes._get_class_value()) if first.classes._get_class_value() else '%s/%s' % (xpath, str(first.tag))
            if not new_xpath in xpath_results:
                if len(root.xpath(first.tag)) > 5:
                    print(new_xpath, first, first.text_content(), list(second.classes))
                    xpath_results.append(new_xpath)
        first_xpath = '%s/%s' % (xpath, str(first.tag))
        find_repeat(first, first_xpath)
    if children_elems:
        second = children_elems[ -1 ]
        second_xpath = '%s/%s' % (xpath, str(second.tag))
        find_repeat(second, second_xpath)
```
